# Dangers of distribution fitting
### An example of fitting distributions to variables that have complex relationships.

In this notebook, we demonstrate the dangers of naively fitting distributions to variables with complex relationships. Specifically, we fit distributions to diastolic and systolic blood pressure values. The key relationship between these two variables is that diastolic blood pressure must always be lower that systolic blood pressure.

```python
# Get raw data, convert to csv, and import
# Import libraries
import os.path
from get_data import get_data
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

# Data locations
raw_dir = "../data/raw/hospital_admissions/5v_cleandf.RData"
processed_dir = "../data/processed/hospital_admissions/"

# If processed data does exist convert raw to csv
if not os.path.exists(processed_dir + "df.csv"):
    get_data.RData_to_csv(raw_dir,processed_dir)
    
# Get data into dataframe
df = pd.read_csv(processed_dir + "df.csv")
df = df[["sbp_last","dbp_last"]]
# Drop nan values
df = df.dropna()
```

Lets have a look at the data. From the histogram plots, both SDP and DBP look to be normally distributed (with some skew).

```python
# Examine data
df.hist(bins=20,figsize=(10,4))
plt.show()
```

Using the SciPy library we can fit skew-normal distributions to both SBP and DBP. This returns a tuple of parameters containing: (skewness, mean, variance)

```python
dist = getattr(stats, 'skewnorm')
p_sbp = dist.fit(df['sbp_last'])
p_dbp = dist.fit(df['dbp_last'])
print(p_sbp)
print(p_dbp)
```

Let's examine the fit, here we plot normalised histograms and fitted distributions.

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_figheight(4); fig.set_figwidth(10)

ax1.hist(df['sbp_last'], bins=50, density=True, alpha=0.6, color='g')
xmin, xmax = ax1.get_xlim(); x = np.linspace(xmin, xmax, 100)
cdf_sbp = stats.skewnorm.pdf(x, p_sbp[0], p_sbp[1], p_sbp[2])
ax1.plot(x, cdf_sbp, 'k', linewidth=2)
ax1.title.set_text('SBP')

ax2.hist(df['dbp_last'], bins=50, density=True, alpha=0.6, color='g')
xmin, xmax = ax2.get_xlim(); x = np.linspace(xmin, xmax, 100)
cdf_dbp = stats.skewnorm.pdf(x,p_dbp[0], p_dbp[1], p_dbp[2])
ax2.plot(x, cdf_dbp, 'k', linewidth=2)
ax2.title.set_text('DBP')
plt.show()
```

```python
plt.figure()
x = np.linspace(0,250, 100)
plt.plot(x, cdf_sbp, linewidth=2, label='SBP')
plt.plot(x, cdf_dbp, linewidth=2, label='DBP')
plt.legend()
plt.show()
```

### Using these fitted distributions to generate new samples

By generating samples from both fitted distributions we can create a synthetic dataset of blood pressures. However, we also observed that a small percentage of the samples are unrealistic, i.e. DBP > SBP. While, in this simple example it would be quite easy to remove these, for more complex relationships between variables this problem becomes increasingly difficult to solve.

```python
# Generate samples
gen_df = pd.DataFrame(
    [stats.skewnorm.rvs(p_sbp[0],p_sbp[1],p_sbp[2],size=10000),
     stats.skewnorm.rvs(p_dbp[0],p_dbp[1],p_dbp[2],size=10000)]).T
gen_df.columns = ['sbp', 'dbp']

# Check for invalid blood pressure readings
gen_df['valid'] = gen_df['sbp'] > gen_df['dbp']

# Percentage of invalid reading, true implies sbp > dbp i.e. valid.
gen_df['valid'].value_counts() /1000
```
